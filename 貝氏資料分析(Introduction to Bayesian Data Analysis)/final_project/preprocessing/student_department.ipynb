# Import

```python
import numpy as np
import pandas as pd
import csv
import datetime
import math
```

# My Library

# Read File

```python
# get all data
data = pd.read_csv('data/data.csv')
```

# Main

```python
students = {}
for i in range(len(data)):
    s_id = data['S_ID'][i]
    department = data['IN_DEPT'][i]
    if s_id not in students:
        if not np.isnan(department):
            students[s_id] = str(int(department)) + 'a'
        else:
            students[s_id] = '0a'

with open('department/student_department.csv', 'w') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['S_ID', 'Department'])
    
    for s_id in students:
        writer.writerow([s_id, students[s_id]])
```
