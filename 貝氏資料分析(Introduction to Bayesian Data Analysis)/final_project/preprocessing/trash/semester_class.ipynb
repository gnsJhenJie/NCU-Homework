# Import

```python
import numpy as np
import pandas as pd
import csv
import datetime
import math
```

# My Library

```python
def class_header():
    return ['S_ID', 'class0', 'class1', 'class2', 'class3', 'class4', 'class5', 'class6', 'class7', 'class8', 'class9', 'class10', 'class11', 'class12', 'class13', 'class14', 'class15', 'class16', 'class17', 'class18', 'class19']
```

# Read File

```python
# get all data
data = pd.read_csv('data/data.csv')
```

# Main

```python
# set all semester variable
semesters = [1011, 1012, 1021, 1022, 1031, 1032, 1041, 1042, 1051, 1052, 1061]

for semester in semesters:
    
    # get specific semester data
    sub_data = data[data['TERM'] == semester]
    
    # write to the target file class_10??.csv
    with open('class_' + str(semester) + '.csv', 'w') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(class_header())
        
        # -----------------------------start to compute the statistic-----------------------------
        students = {}
        
        #record each score
        for i in range(len(sub_data)):
            s_id = sub_data['S_ID'].values[i]
            the_class = sub_data['CNO'].values[i] + str(sub_data['CLS'].values[i])
            if s_id not in students:
                students[s_id] = [the_class]
            else:
                students[s_id].append(the_class)
                
        # average all score in each S_ID
        for s_id in students:
            writer.writerow([s_id] + students[s_id])
    
    print(semester, 'ok')
```
